# Bloom Filter - Demonstration and Testing

This notebook shows how to use our custom-made data structure, `BloomFilter`. Certainly, according to the requirements of the project, the core logic is implemented in a separate module(`bloom_filter.py`). That way, the module can be easily imported here to be logically tested and demonstrated, and be used by our `benchmark.py` script for stress testing on the HPC cluster on a huge scale.

Here we will show how to perform basic insertions, membership checks and deal with expected outcomes.

```python
# Import our custom Bloom Filter module
from bloom_filter import BloomFilter
```

```python
# Initialize the Bloom Filter for a small dataset (20 items) with a 5% false positive rate
bf = BloomFilter(expected_items=20, fp_rate=0.05)

# 1. Insertion Test
fruits_to_add = ["apple", "banana", "cherry", "strawberry", "orange"]

print("--- Inserting Items into Bloom Filter ---")
for fruit in fruits_to_add:
    bf.add(fruit)
    print(f"Inserted: {fruit}")
```

```python
print("--- Checking Membership ---")
print(f"Is 'apple' in the filter?       -> {bf.check('apple')} (Expected: True)")
print(f"Is 'mango' in the filter?       -> {bf.check('strawberry')} (Expected: True)")

print("\n--- Checking Non-Existing Items ---")
print(f"Is 'grape' in the filter?       -> {bf.check('grape')} (Expected: False)")
print(f"Is 'watermelon' in the filter?  -> {bf.check('watermelon')} (Expected: False)")
```
